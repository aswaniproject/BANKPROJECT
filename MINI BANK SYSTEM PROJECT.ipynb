```python
class BankAccount:
    def __init__(self, account_holder, password, balance=0):
        self.account_holder = account_holder
        self.password = password
        self.balance = balance
        self.transactions = []  # Store transactions
    
    def login(self, input_password):
        if self.password == input_password:
            print(f"Login successful. Welcome, {self.account_holder}.")
            return True
        else:
            print("Invalid password. Please try again.")
            return False
    
    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            self.transactions.append(f"Deposited {amount}")
            print(f"Deposited: {amount}. New balance: {self.balance}")
        else:
            print("Deposit amount must be positive.")
    
    def withdraw(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount
            self.transactions.append(f"Withdrew {amount}")
            print(f"Withdrawn: {amount}. New balance: {self.balance}")
        else:
            print("Insufficient balance or invalid amount.")
    
    def check_balance(self):
        print(f"Current balance: {self.balance}")

    def add_interest(self, interest_rate):
        """Add interest to the balance."""
        interest = self.balance * (interest_rate / 100)
        self.balance += interest
        self.transactions.append(f"Interest added: {interest}")
        print(f"Interest of {interest} added. New balance is: {self.balance}")
    
    def change_password(self, old_password):
        """Allow the user to change their password."""
        if old_password == self.password:
            new_password = input("Enter new password: ")
            self.password = new_password
            print("Password changed successfully.")
        else:
            print("Incorrect old password. Try again.")
    
    def show_transactions(self):
        """Display transaction history."""
        print("Transaction History:")
        for transaction in self.transactions:
            print(transaction)

# Main Program
def main():
    print("Welcome to the Banking System.")
    account = BankAccount("ASWANIS", "ACHU123", 4000)
    
    while True:
        input_password = input("Enter your password: ")
        if account.login(input_password):
            break
    
    while True:
        print("\nOptions:")
        print("1. Deposit")
        print("2. Withdraw")
        print("3. Check Balance")
        print("4. Add Interest")
        print("5. Change Password")
        print("6. Show Transactions")
        print("7. Exit")
        
        choice = input("Choose an option (1-7): ")
        
        if choice == "1":
            amount = float(input("Enter amount to deposit: "))
            account.deposit(amount)
        elif choice == "2":
            amount = float(input("Enter amount to withdraw: "))
            account.withdraw(amount)
        elif choice == "3":
            account.check_balance()
        elif choice == "4":
            interest_rate = float(input("Enter interest rate (%): "))
            account.add_interest(interest_rate)
        elif choice == "5":
            old_password = input("Enter your old password: ")
            account.change_password(old_password)
        elif choice == "6":
            account.show_transactions()
        elif choice == "7":
            print("Thank you for using the banking system.")
            break
        else:
            print("Invalid option. Please choose again.")

if __name__ == "__main__":
    main()
```
